```python
import pandas as pd
from sklearn import datasets
```

## 1. Load the Data

```python
data = datasets.load_iris(as_frame=True)
df = data.frame
```

## 2. Examine the Data

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['target'].unique()
```

## 3. Clean the Data

```python
df.columns
```

```python
df = df.rename(columns=lambda x: '_'.join(x.split()[:2]))
df.head()
```

## 4. Analyze the Data

```python
df.corr()
```

```python
f = pd.plotting.scatter_matrix(df, figsize=(14, 10))
```

```python
b = df.boxplot(
    column=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], 
    by='target', 
    figsize=(12, 10)
)
```

## 5. Export the Data

```python
df.to_csv('iris_df.csv', index=False)
```
